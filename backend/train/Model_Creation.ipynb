```python
from sklearn.metrics import classification_report, f1_score
from tabulate import tabulate

# Initialize results list
results = []

# Dictionary of models and their predictions
models = {
    'RandomForestClassifier': y_pred,
    'GradientBoostingClassifier': y_pred1,
    'XGBClassifier': y_pred2,
    'LogisticRegression': y_pred3,
    'DecisionTreeClassifier': y_pred4,
    'KNeighborsClassifier': y_pred5
}


# Evaluate each model
for model_name, y_pred in models.items():

    # Compute weighted F1 score
    f1 = f1_score(y_test, y_pred, average='weighted')
    acc = accuracy_score(y_test,y_pred)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    # Calculate specificity and sensitivity
    specificity = tn / (tn + fp)
    sensitivity = tp / (tp + fn)


    # Append the results
    results.append((model_name, f1, acc, specificity, sensitivity))

# Display results as a table
print(tabulate(results, headers=["Model", "F1 Score", "Accuracy Score", "Specificity", "Sensitivity"], tablefmt="grid"))
```
